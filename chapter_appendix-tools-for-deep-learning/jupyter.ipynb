```python
b= [1,5,8,10.5,20.6,19.95, [1,2,3,5.5,6.5,7.5]]

sum_even=0
sum_float=0
for i in b:
  if type(i)==list:
    for j in b[-1]:
      if type(j)==int:
        sum_even+=j
      else:
        sum_float+=j

  elif type(i)!=list:
    for k in b[0:len(b)-1]:
      if type(k)==int:
        sum_even+=0
      else:
        sum_float+=0

print(sum_even)
print(sum_float)
```
