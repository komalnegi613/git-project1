```python
def sum_of_two_digit(num):
    # Check if the number is two digits
    if num < 10 or num > 99:
        return "Please enter a two-digit number"

    tens = num // 10      # first digit
    ones = num % 10       # second digit

    return tens + ones
```

```python
sum_of_two_digit(62)
```
